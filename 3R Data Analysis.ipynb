```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('3R COVER DEFECTS.xlsx')
df['Tunnel/Viaduct'] = df['Tunnel/Viaduct'].fillna('Viaduct')
df.head()
```

```python
colors = list(df['Tunnel/Viaduct'].replace({'Viaduct':'blue','Tunnel':'brown'}))
```

```python
sectors = list(df['TrackSector'])
#for index in range(0,len(sectors)):
#    if index%2 ==0:
#        sectors[index] = ''
```

```python
# create a dataset
count = df['Defect Count']
x_pos = np.arange(len(sectors))

# Create title and labels
plt.title('3R Cover Defects in NSEWL (2018-2020)',fontsize=30)
plt.xlabel('Sectors',fontsize=15)
plt.ylabel('Defect Count',fontsize=15)

# Create bars with different colors
plt.bar(x_pos, count, color=colors)

# Create names on the x-axis
plt.xticks(x_pos, sectors,rotation=90,fontsize= 10)

#Create legend
legend_colors = {'Viaduct':'blue','Tunnel':'brown'}
labels = list(legend_colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=legend_colors[label]) for label in labels]
plt.legend(handles, labels, prop={'size':15})

# Show graph
plt.rcParams["figure.figsize"] = (30,30)
plt.show()
```

```python
df = pd.read_excel('3R Nylon Bolt Defects.xlsx')
df['Tunnel/Viaduct'] = df['Tunnel/Viaduct'].fillna('Viaduct')
df = df.sort_values(by=['Defect Count'],ascending=[False])
```

```python
colors = list(df['Tunnel/Viaduct'].replace({'Viaduct':'blue','Tunnel':'brown'}))
sectors = list(df['TrackSector'])
```
